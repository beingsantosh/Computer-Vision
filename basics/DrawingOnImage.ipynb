## Drawing on a Image


```python

import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
blank_img = np.zeros((512,512,3), dtype= np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img) # all is zero
```

```python
cv2.rectangle(blank_img, pt1 = (100,100), pt2 = (450,450), color = (0,256,0), thickness = 5)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(256,256), radius = 100, color= (256,256,256), thickness = 10)

plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(400,200), radius = 50, color= (0,256,256), thickness = -1)

plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0,0), pt2 =(512,512),color=(256,0,0),thickness = 3)
plt.imshow(blank_img)
```

```python
# text on image

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='add any text here', org = (10, 500), fontFace = font, 
            fontScale= 1, color = (256,0,256), 
            thickness = 4, lineType = cv2.LINE_AA )
plt.imshow(blank_img)
```

```python
# custom polygons
blank_img1 = np.zeros((512,512,3), dtype= np.int32)
```

```python
plt.imshow(blank_img1, cmap= 'gray')
```

```python
vertices = np.array([[100,200], [200,50], [400,300], [150,400] ], dtype= np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
pt=vertices.reshape(-1,1,2)
```

```python
pt.shape
```

```python
pt
```

```python
cv2.polylines(blank_img1, [pt], isClosed= True, color = (255,0,0), thickness=4)
plt.imshow(blank_img1)
```
